```python
import requests

file_names = ["/internet-movie-database/the_shawshank_redemption_summaries.txt",
              "/nba/teams_rowbased.json",
              "/nba/teams_columnbased.json",
              "/internet-movie-database/movies.csv"]
for file_name in file_names:
    file_url = f"https://raw.githubusercontent.com/datainpoint/classroom-hahow-pythonfiftyplus/main/data{file_name}"
    r = requests.get(file_url)
    file_name_split = file_name.split("/") 
    with open(file_name_split[-1], "w") as f:
        f.write(r.text)
```
