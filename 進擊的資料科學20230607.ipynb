```python
import pandas as pd

movies = pd.read_csv("movies.csv")
print(movies.shape)
print(type(movies))
```

```python
runtime_hours = movies["runtime"] // 60 # floor divide
runtime_mins = movies["runtime"] % 60 # modulo
```

```python
runtime_hours = runtime_hours.astype(str)
runtime_mins = runtime_mins.astype(str)
runtime_hours.str.cat(runtime_mins, sep=":")
```

```python
def runtime_format(x: int) -> str:
    hours = str(x // 60)
    mins = str(x % 60)
    hours_mins = f"{hours.zfill(2)}:{mins.zfill(2)}"
    return hours_mins

runtime_format(142)
runtime_format(175)
runtime_format(122)
```

```python
movies["runtime"].map(runtime_format)
```

```python
import numpy as np

def get_director_middle_name(x: str) -> str:
    if len(x) == 2:
        return np.nan
    elif len(x) == 3:
        return x[1]
    elif len(x) == 4:
        return f"{x[1]} {x[2]}"
director_first_names = director_str_split.map(lambda x: x[0])
director_last_names = director_str_split.map(lambda x: x[-1])
director_middle_names = director_str_split.map(get_director_middle_name)
```

```python
towns = []
request_url = "https://emap.pcsc.com.tw/EMapSDK.aspx"
for city_id in range(1, 4): # (1, 26)
    request_url = "https://emap.pcsc.com.tw/EMapSDK.aspx"
    str_city_id = str(city_id)
    form_data = {
        "commandid": "GetTown",
        "cityid": f"{str_city_id.zfill(2)}"
    }
    r = requests.post(request_url, data=form_data)
    tree = ET.fromstring(r.text)
    town_names = [e.text for e in tree.findall(".//TownName")]
    towns += town_names
print(towns)
```
